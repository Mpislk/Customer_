```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
```

```python
data1 = pd.read_csv('dim_customers.csv')
data2 = pd.read_csv('fact_spends.csv')
```

```python
df = (data2.groupby('customer_id')['spend'].sum()/6).round(0).reset_index()
```

```python
df2 = pd.merge(data1,df)
df2['income_util_ratio'] = (100*df2['spend']/df2['avg_income']).round(0)
```
